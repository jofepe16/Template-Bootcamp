<span style="color: #007BFF; font-size: 24px; font-weight: bold; border: 2px solid red; padding: 10px; background-color: #EFEFEF; display: block; width: 100%;"> 3. Gráficos Relacionales</span>

Los gráficos relacionales son herramientas fundamentales en el análisis de datos que nos permiten explorar y visualizar cómo se relacionan dos o más variables entre sí. Estos gráficos son especialmente útiles para entender las asociaciones y patrones que pueden no ser evidentes al observar los datos en forma tabular. Aquí te explico el contexto y la utilidad de los gráficos relacionales:

#### **1. Objetivo de los Gráficos Relacionales:**

Los gráficos relacionales se utilizan para visualizar la relación entre variables, permitiendo una comprensión más profunda de cómo una variable afecta o se relaciona con otra. Son útiles para identificar tendencias, correlaciones, y patrones en los datos. 

#### **2. Tipos de Gráficos Relacionales:**

- **Gráfico de Dispersión (Scatter Plot):** Muestra la relación entre dos variables numéricas. Cada punto en el gráfico representa una observación en el conjunto de datos, con la posición en el eje X e Y correspondiente a los valores de las dos variables.
  
- **Gráfico de Líneas (Line Plot):** Ideal para visualizar la relación entre variables cuando una de ellas representa una medida de tiempo. Permite observar tendencias y cambios a lo largo del tiempo.
  
- **Gráfico de Relación (relplot):** Una función de Seaborn que permite crear gráficos de dispersión o de líneas con facilidad. La función <code>relplot()</code> es especialmente útil para crear gráficos que muestran cómo una variable se relaciona con otra, con opciones para ajustar el tipo de gráfico según el contexto.

#### **3. Aplicaciones Comunes:**

- **Análisis de Tendencias:** Los gráficos de líneas son útiles para analizar cómo las variables cambian a lo largo del tiempo. Por ejemplo, podrías usar un gráfico de líneas para visualizar la evolución del número de pasajeros a bordo del Titanic a lo largo de las distintas fechas de embarque.

- **Detección de Correlaciones:** Los gráficos de dispersión ayudan a identificar la fuerza y dirección de una relación entre dos variables. Por ejemplo, podrías analizar la relación entre la tarifa pagada y la edad de los pasajeros para ver si existe alguna tendencia.

- **Comparación de Subgrupos:** Los gráficos relacionales también permiten comparar diferentes subgrupos dentro de los datos. Por ejemplo, podrías comparar la relación entre la clase socioeconómica y la supervivencia de los pasajeros, usando diferentes colores o marcadores para distinguir entre clases.

#### **4. Ventajas:**

- **Visualización Clara:** Ofrecen una representación visual clara de las relaciones entre variables, facilitando la interpretación de datos complejos.
  
- **Identificación de Patrones:** Ayudan a identificar patrones, tendencias y anomalías que pueden no ser evidentes en los datos brutos.

- **Facilidad de Interpretación:** Los gráficos relacionales permiten una interpretación intuitiva de las relaciones entre variables, apoyando el análisis exploratorio de datos.

#### **5. Herramientas:**

- **Seaborn:** Una biblioteca de Python que proporciona una interfaz fácil de usar para crear gráficos relacionales con <code>relplot()</code>. Permite ajustar el tipo de gráfico y personalizar la visualización según las necesidades del análisis.

En resumen, los gráficos relacionales son esenciales para explorar y entender la interrelación entre variables, facilitando la interpretación de datos y la identificación de patrones importantes en el análisis.

<span style="color: #007BFF; font-size: 24px; font-weight: bold; border-left:3px solid red; border-right:3px solid red; padding: 10px;">
    3.3. lineplot
</span


---

**Lineplot** es una forma de trazar un gráfico de líneas con la posibilidad de varios agrupamientos semánticos.

La relación entre **x** y **y** se puede mostrar para diferentes subconjuntos de los datos utilizando los parámetros **hue**, **size** y **style**. Estos parámetros controlan qué semánticas visuales se utilizan para identificar los diferentes subconjuntos. Es posible mostrar hasta tres dimensiones independientemente utilizando los tres tipos semánticos, pero este estilo de gráfico puede ser difícil de interpretar y a menudo es ineficaz. Utilizar semánticas redundantes (es decir, tanto **hue** como **style** para la misma variable) puede ser útil para hacer los gráficos más accesibles.

El tratamiento predeterminado del **hue** (y, en menor medida, **size**), si está presente, depende de si la variable se infiere que representa datos **“numéricos” o “categóricos”**. En particular, las variables numéricas se representan con un mapa de colores secuencial por defecto, y las entradas de la leyenda muestran **“ticks”** regulares con valores que pueden o no existir en los datos. Este comportamiento se puede controlar a través de varios parámetros.

Por defecto, el gráfico agrega sobre múltiples valores de **y** en cada valor de **x** y muestra una estimación de la tendencia central y un intervalo de confianza para esa estimación.



```python
plt.figure(figsize=(8,8))
sns.lineplot(x="Age", y="Fare", hue="Sex", style="Sex", data=full_data)
plt.show()
```

<span style="color: #007BFF; font-size: 24px; font-weight: bold; border-left:3px solid red; border-right:3px solid red; padding: 10px;">
    4.1. Barplot
</span

---

**`barplot()`** muestra la relación entre una variable categórica y una variable continua. Los datos se representan en barras rectangulares donde la longitud de la barra representa la proporción de los datos en esa categoría. El **Bar plot** representa la estimación de la **tendencia central**.


**Nota:** No confundir **Bar plot** con **Histograma**. Aquí están las principales diferencias entre ellos:

- **Barplot:**
  - **Variables:** Utiliza una variable categórica en el eje x y una variable continua en el eje y.
  - **Propósito:** Muestra la relación entre categorías y su valor promedio o alguna medida de tendencia central.
  - **Ejemplo:** Comparar el promedio de tarifas (`Fare`) pagadas por diferentes clases socioeconómicas (`Pclass`).

- **Histograma:**
  - **Variables:** Utiliza una sola variable continua en el eje x y muestra la frecuencia de ocurrencia en el eje y.
  - **Propósito:** Muestra la distribución de una variable continua, es decir, cómo se distribuyen los valores a lo largo de su rango.
  - **Ejemplo:** Mostrar cómo se distribuyen las edades (`Age`) de los pasajeros.

El **Bar plot** es útil para comparar diferentes categorías, mientras que el **Histograma** es útil para entender la distribución de una sola variable continua.


```python
plt.figure(figsize=(8,8))
sns.barplot(x="Sex", y="Survived", hue="Pclass", data=full_data)
plt.show()
```

   <span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
        4.2. countplot
    </span>
   <br><br>

   <p>
        A special case in <strong>barplot</strong> is to show the number of observations in each category rather than computing a statistic for a second variable. For this, we use <code>countplot()</code>.
    </p>

</div>

```python
sns.countplot(x="Survived", data=full_data, palette="Blues");
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(7, 9), sharex=True, sharey=True)

# Configuración de colores
palette = ["#f9dbbd", "#DA627D", "#450920"]  # Paleta azul para las clases

# Primer gráfico: Número absoluto de pasajeros por Sexo y Clase
sns.countplot(x="Sex", hue="Pclass", data=full_data, palette=palette, ax=axs[0])
axs[0].set_title("Número absoluto de pasajeros por Sexo y Clase")
axs[0].set_xlabel("Sexo")
axs[0].set_ylabel("Pasajeros")
axs[0].legend(title="Clase")

# Segundo gráfico: Número absoluto de sobrevivientes por Sexo y Clase
sns.countplot(x="Sex", hue="Pclass", data=full_data[full_data.Survived==1], palette=palette, ax=axs[1])
axs[1].set_title("Número absoluto de sobrevivientes por Sexo y Clase")
axs[1].set_xlabel("Sexo")
axs[1].set_ylabel("Sobrevivientes")
axs[1].legend(title="Clase")

# Ajuste de diseño para evitar superposiciones
plt.tight_layout()
plt.show()
```

<strong>El gráfico muestra que el número de pasajeros en la tercera clase es mayor que en la primera y segunda clase.</strong>

**Como podemos ver, sobrevivieron más mujeres que hombres.**

       <br><br>

Los diagramas de caja suelen tener líneas verticales que se extienden desde las cajas, llamadas <strong>bigotes</strong>. Estos bigotes indican la variabilidad fuera de los cuartiles superior e inferior, por lo que los Boxplots también se conocen como <strong>diagrama de caja y bigotes</strong>. Los valores atípicos en los datos se representan como puntos individuales.
    </p>

</div>


```python
plt.figure(figsize=(5,4))
sns.boxplot(y="Age", data=full_data)
plt.show()
```

   <br><br>

        <strong>Los gráficos de puntos</strong> sirven para lo mismo que los <strong>gráficos de barras</strong>, pero en un estilo diferente. En lugar de una barra completa, el valor de la estimación se representa mediante un punto a una cierta altura en el otro eje.
    </p>

</div>

```python
plt.subplots(figsize=(8, 8))
sns.pointplot(x="Sex", y="Survived", hue="Pclass", data=full_data)
plt.show()
```

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

<strong>Podemos ver el número promedio de sobrevivientes de hombres y mujeres en cada clase. A partir del gráfico, podemos entender que una mayor proporsión de mujeres sobrevivió en comparación con los hombres. Tanto en hombres como en mujeres, la mayor cantidad de sobrevivientes proviene de la primera clase.</strong>
    
</div>

# <span style="color: #007BFF; font-size: 24px; font-weight: bold; border: 2px solid #007BFF; padding: 10px; background-color: #EFEFEF; display: block; width: 100%;"> 5. Gráficos de Regresión</span>

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">
   <p>
        La mayoría de las veces, utilizamos conjuntos de datos que contienen múltiples variables cuantitativas, y el objetivo de un análisis a menudo es relacionar esas variables entre sí. Esto se puede hacer mediante las <em>líneas de regresión</em>.
    </p>
   <p>
        Mientras construimos los modelos de regresión, a menudo verificamos la <em>multicolinealidad</em>, donde tenemos que observar la correlación entre todas las combinaciones de variables continuas y tomar las medidas necesarias para eliminar la multicolinealidad si existe.
    </p>

   <span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
        5.1. lmplot
    </span>

</div>

```python
sns.lmplot(x="Age", y="Fare", data=full_data)
plt.show()
```

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

<span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">5.2. regplot</span> <br><br>
    
`regplot` acepta las variables x e y en una variedad de formatos, incluyendo arreglos simples de numpy, objetos Series de pandas, o como referencias a variables en un DataFrame de pandas.
    
</div>

```python
from scipy.stats import pearsonr

plt.subplots(figsize=(10, 10))
sns.regplot(x="Age", y="Fare", data=full_data)
plt.show()

# Eliminar filas con valores NaN en las columnas 'Age' y 'Fare'
cleaned_data = full_data[['Age', 'Fare']].dropna()
# Calcular el coeficiente de correlación de Pearson
correlation_coefficient, _ = pearsonr(cleaned_data['Age'], cleaned_data['Fare'])
# Imprimir el coeficiente de correlación
print(f"Coeficiente de correlación de Pearson entre Edad y Tarifa: {correlation_coefficient:.2f}")
```

```python
plt.subplots(figsize=(10, 10))
sns.regplot(x="Fare", y="Fare", data=full_data)
plt.show()
```

# <span style="color: #007BFF; font-size: 24px; font-weight: bold; border: 2px solid #007BFF; padding: 10px; background-color: #EFEFEF; display: block; width: 100%;"> 6. Gráficos de Matriz</span><br>

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">

<span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
    6.1. heatmap
</span> <br><br>
    Visualizar datos con <strong>heatmaps</strong> es una excelente manera de realizar un análisis exploratorio de datos cuando se tiene un conjunto de datos con múltiples variables. Los heatmaps pueden revelar patrones generales en el conjunto de datos de manera instantánea. Y es muy fácil crear heatmaps atractivos con la biblioteca Seaborn en Python.

Ahora vamos a trazar la matriz de correlación de nuestros datos con un heatmap.


```python
# Calcular la matriz de correlación
correlation_matrix = full_data.corr()

# Crear el heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Matriz de Correlación")
plt.show()
```

```python
# Selecciona solo columnas numéricas
numeric_data = full_data.select_dtypes(include=['number'])

# Calcular la matriz de correlación
correlation_matrix = numeric_data.corr()

# Crear el heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(correlation_matrix, cmap="YlGnBu", annot=True, fmt=".2f")
plt.title("Matriz de Correlación")
plt.show()
```

**Observamos que las correlaciones entre las variables son débiles.**

# <span style="color: #007BFF; font-size: 24px; font-weight: bold; border: 2px solid #007BFF; padding: 10px; background-color: #EFEFEF; display: block; width: 100%;"> 7. Multi-plot grids</span>
<br>

<span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
    7.1. Facet grids
</span>
<br><br>

<div style="font-family: 'Arial', sans-serif; font-size: 16px; line-height: 1.6; color: #333;">
   <p>
    Una técnica útil para explorar los datos a trav{es de varias dimensiones es dibujar múltiples instancias del mismo gráfico en diferentes subconjuntos de tu conjunto de datos. Esta técnica se denomina comúnmente <em>“rejilla”</em> o <em>“paneles”</em>, y está relacionada con la idea de <em>“múltiples pequeños”</em>. Para utilizar estas funciones, tus datos deben estar en un DataFrame de Pandas.
 </p>

   <p>
        <strong>La rejilla de paneles</strong> forma una matriz de paneles definidos por filas y columnas al dividir las variables. Debido a los paneles, un solo gráfico parece ser múltiples gráficos. Es muy útil para analizar todas las combinaciones en dos variables discretas.
    </p>
   <p>
        El objeto <strong>FacetGrid</strong> toma un DataFrame como entrada y los nombres de las variables que formarán las dimensiones de fila, columna o tono de la rejilla. Las variables deben ser categóricas, y los datos en cada nivel de la variable se utilizarán para un panel a lo largo de ese eje.
    </p>
   <p>
        La ventaja de usar Facet es que podemos ingresar otra variable en el gráfico. Podemos crear muchos paneles de columna y alinearlos con las filas de la rejilla.
    </p>
</div>

#### FacetGrid.map

El enfoque principal para visualizar datos en esta cuadrícula es con el método `FacetGrid.map()`

```python
# initialize the FacetGrid object
g = sns.FacetGrid(full_data, col='Survived', row='Pclass')

g.map(plt.hist, 'Age')
g.add_legend()
plt.show()
```

# <span style="color: #007BFF; font-size: 24px; font-weight: bold; border: 2px solid #007BFF; padding: 10px; background-color: #EFEFEF; display: block; width: 100%;"> 8. Otras visualizaciones</span>
<br>

<span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
    Diagramas de torta
</span>
<br><br>

Para entender mejor la composición de los pasajeros en el conjunto de datos, se puede utilizar un gráfico de torta (diagrama circular). Este gráfico visualiza la proporción de cada categoría de género dentro del total de pasajeros.

```python
import matplotlib.pyplot as plt

# Crear el gráfico de torta para la distribución de género
gender_counts = full_data['Sex'].value_counts()
plt.figure(figsize=(4, 4))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Distribución de Género de los Pasajeros')
plt.show()
```

<span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
    Gráfico de Violin (Violin Plot):
</span>
<br><br>

Similar a un box plot, pero también muestra la densidad de la distribución de la variable.
Ejemplo: Mostrar la distribución del Age por clase (Pclass).

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x="Pclass", y="Age", data=full_data, palette="Set1")
plt.title('Distribución de Edad por Clase')
plt.xlabel('Clase')
plt.ylabel('Edad')
plt.show()
```

<span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
    Gráfico de Pareto:
</span>
<br><br>

Uso: Para visualizar la frecuencia acumulativa de eventos o categorías. Es útil para entender la importancia relativa de diferentes categorías.
Ejemplo: Ver la distribución acumulativa de embarque o puerto de partida (Embarked).

```python
import pandas as pd

survival_counts = full_data['Embarked'].value_counts().sort_index()
survival_percent = survival_counts / survival_counts.sum() * 100
cumulative_percent = survival_percent.cumsum()

plt.figure(figsize=(10, 6))
plt.bar(survival_counts.index, survival_percent, color='lightblue', label='Porcentaje')
plt.plot(survival_counts.index, cumulative_percent, color='red', marker='o', label='Porcentaje Acumulado')
plt.title('Distribución de Embarque y Porcentaje Acumulado')
plt.xlabel('Embarque')
plt.ylabel('Porcentaje')
plt.legend()
plt.show()
```

<span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
    Pair Plot:
</span>
<br><br>

Uso: Para explorar las relaciones entre todas las combinaciones de variables numéricas en un conjunto de datos.
Ejemplo: Analizar las relaciones entre Age, Fare y Pclass.

```python
sns.pairplot(full_data[['Age', 'Fare', 'Pclass']].dropna())
plt.show()
```

<span style="color: #007BFF; font-size: 24px; font-weight: bold;  border-left:3px solid red; border-right:3px solid red; padding: 10px;">
    Gráfico de Contingencia (Heatmap de Variables Categóricas):
</span>
<br><br>

Uso: Para visualizar la relación entre dos variables categóricas.
Ejemplo: Mostrar la relación entre Pclass y Survived.

```python
contingency_table = pd.crosstab(full_data['Pclass'], full_data['Survived'])
sns.heatmap(contingency_table, annot=True, cmap="YlGnBu", fmt="d")
plt.title('Relación entre Clase y Supervivencia')
plt.xlabel('Sobrevivió')
plt.ylabel('Clase')
plt.show()
```
